```python
test_1 = test.groupby('variant').agg({
    'user_id': 'nunique',
    'cta_clicked': 'sum',
    'signup_completed': 'sum',
    'recommendation_cards_clicked': 'sum',
    'bounced': 'sum'
})
print(test_1)
```

```python
test_perc = round(test.groupby('variant').agg({
    'user_id': 'nunique',
    'cta_clicked': 'sum',
    'signup_completed': 'sum',
    'recommendation_cards_clicked': 'sum',
    'bounced': 'sum'
}) * 100 / (test.shape[0]/2),2)

test_perc
```

```python
z_stat, p_value = sm.stats.proportions_ztest(test_1['cta_clicked'], test_1['user_id'])

print(f'CTA_clicked results:')
print("\nZ-statistic:", z_stat)
print("P-value:", p_value)
if p_value > 0.05:
         print('Results are not statistically significant.')
else:
         print(' Results are statistically significant.')

print(f'Absolute difference: {abs(test_1['cta_clicked'].iloc[0] - test_1['cta_clicked'].iloc[1])} , {round(abs(test_perc['cta_clicked'].iloc[0] -  test_perc['cta_clicked'].iloc[1]),2)}%')
print(f'Relative difference: {round((test_perc['cta_clicked'].iloc[1] - test_perc['cta_clicked'].iloc[0]) * 100 / test_perc['cta_clicked'].iloc[0],2)} %')
```

```python
z_stat, p_value = sm.stats.proportions_ztest(test_1['signup_completed'], test_1['user_id'])

print(f'Signup Rate results:')
print("\nZ-statistic:", z_stat)
print("P-value:", p_value)
if p_value > 0.05:
         print('Results are not statistically significant.')
else:
         print(' Results are statistically significant.')


print(f'Absolute difference: {abs(test_1['signup_completed'].iloc[0] - test_1['signup_completed'].iloc[1])} , {round(abs(test_perc['signup_completed'].iloc[0] -  test_perc['signup_completed'].iloc[1]),2)}%')
print(f'Relative difference: {round((test_perc['signup_completed'].iloc[1] - test_perc['signup_completed'].iloc[0]) * 100 / test_perc['signup_completed'].iloc[0],2)} %')
```

```python
z_stat, p_value = sm.stats.proportions_ztest(test_1['bounced'], test_1['user_id'])

print(f'Bounced Rate results:')
print("\nZ-statistic:", z_stat)
print("P-value:", p_value)
if p_value > 0.05:
         print('Results are not statistically significant.')
else:
         print(' Results are statistically significant.')


print(f'Absolute difference: {abs(test_1['bounced'].iloc[0] - test_1['bounced'].iloc[1])} , {round(abs(test_perc['bounced'].iloc[0] -  test_perc['bounced'].iloc[1]),2)}%')
print(f'Relative difference: {round((test_perc['bounced'].iloc[1] - test_perc['bounced'].iloc[0]) * 100 / test_perc['bounced'].iloc[0],2)} %')
```

Despite positive changes in metrics, most results are not statistically significant because sample groups are not large enough to support a 3% lift instead of a 5% one.
However, there is positive dynamics. Therefore, it is recommended to continutue gathering data. To detect 3% uplift in CTA rate the test should be run for 10 days more.
